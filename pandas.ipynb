```python
sales=pd.read_csv("sales_data.csv")
```

```python
sales.isnull().values.any()
```

```python
sales.describe()
```

```python
sales.dropna()
```

```python
sales.dropna(axis=1)
```

```python
sales.duplicated()
```

```python
sales.drop_duplicates()
```

```python
sales['total sales']=sales['quantity']*sales['price_per_unit']
```

```python
sales.head()
```

```python
sales['date']=pd.to_datetime(sales['date'])
```

```python
sales['month']=sales['date'].dt.month
sales['year']=sales['date'].dt.year
```

```python
sales.head()
```

```python
total_sales_per_product=sales.groupby('product')['total sales'].transform('sum')
```

```python
sales['total_sales_per_product']=total_sales_per_product
```

```python
total_sales_per_store=sales.groupby('store')['total sales'].transform('sum')
```

```python
sales['total_sales_per_store']=total_sales_per_store
```

```python
print(sales)
```

```python
customers=pd.read_csv('customers.csv')
```

```python
merged_df=pd.merge(sales,customers,on='customer_id',how='inner')
```

```python
print(merged_df)
```

```python
merged_df.to_csv('merged_df.csv')
```

```python
customer_spending=merged_df.groupby('customer_id')['total sales'].sum()
```

```python
top_5_customers=customer_spending.sort_values(ascending=False).head(5)
```

```python
print(top_5_customers)
```

```python
pivot_table=pd.pivot_table(merged_df,values='total sales',index='month',columns='store',aggfunc='sum',fill_value=0)
```

```python
print(pivot_table)
```
